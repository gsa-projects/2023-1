```python
import random
hangeuls = list("가나다라마바사아자차카타파하")

with open("info.txt", "w") as file:
    for i in range(10000): 
        name = random.choice(hangeuls) + random.choice(hangeuls)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)
        file.write(f"{name} {weight} {height}\n")
```

```python
with open("info.txt") as file:
    for line in file:
        print(line)
        (name, weight, height) = line.strip().split(" ")
        
        # if (name == None) or (weight == None) or (height == None): 이랑 같은 말
        if (not name) or (not weight) or (not height):
            continue
        
        bmi = int(weight) / (int(height) * int(height))
        result = ""
        
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상 체중"
        else:
            result = "저체중"
            
        print(f"이름: {name}, 체중: {weight}, 키: {height}, BMI: {bmi}, 결과: {result}")
        print()
```
